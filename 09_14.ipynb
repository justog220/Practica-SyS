```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve
from scipy.linalg import  norm
from scipy.fft import fft, ifft
```

**Ejercicio 2:**

Dado el sistema 6y[n] − 4y[n − 1] + 5y[n − 2] = x[n] − 2x[n − 1] + x[n − 2]

inicialmente en reposo, obtenga la respuesta al escalón unitario mediante la ecuación en diferencias y luego compárela con la calculada mediante la sumatoria de convolución, para lo que deberá encontrar previamente su respuesta al impulso.

Salida frente al escalón unitario
- fm =100 Hz
- o <= t <= 1

1). Ciclo for

2). Convolución Lineal

  2.1). Obtener h(t) --> poner el delta de dirac a la entrada
  
  2.2). Calcular y(t)


```python
#Creamos la señal
Tm = 0.01
fm = 1/Tm #100Hz
t = np.arange(0,1,1/fm)
N = len(t)

#Salida
y = np.zeros(N)

#Entrada
U = np.ones(N) #deltas de dirac

#y[n]=(1/6) * ( x[n] - 2*x[n-1] + x[n-2] + 4*y[n-1] - 5*y[n-2] )
y[0] = (1/6)*U[0]
y[1] = (1/6)*( U[1] - 2*U[1-1] + 4*y[1-1] )

for i in range(2,N):
    y[i] = (1/6) * (U[i] - 2*U[i-1] + U[i-2] + 4*y[i-1] - 5*y[i-2])

#y[N-2] = (1/6) * ( U[N-2] - 2*U[N-3] + U[N-4] + 4*y[N-2] - 5*y[N-4] )
#y[N-1] = (1/6) * ( U[N-1] - 2*U[N-2] + U[N-3] + 4*y[N-1] - 5*y[N-3] )


#-------------------Grafico las señales
plt.subplots(2,1, figsize=(14,7))
#--------------------------
plt.subplot(2,1,1)
#plt.plot(t,U)
plt.stem(t, U, use_line_collection=True)
plt.title("Señal de entrada U")
plt.ylabel("Amplitud")
#--------------------------
plt.subplot(2,1,2)
plt.plot(t,y)
plt.title("Señal de salida y")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")

plt.show()
```

```python
#-------------------Calculo el espectro
S_U = fft(U)
Sabs_U = np.abs(S_U)
f_U = np.arange(0,fm/2,fm/N)

S = fft(y)
Sabs = np.abs(S)
f = np.arange(0,fm/2,fm/N)

#------------------Grafico los espectros
plt.subplots(2,1, figsize=(14,7))
#---------------------------------
plt.subplot(2,1,1)
plt.title("Espectro de entrada x")
plt.ylabel("|S[k]|")
plt.stem(f_U, Sabs_U[0:int(N/2)], use_line_collection=True)
#---------------------------------
plt.subplot(2,1,2)
plt.title("Espectro de salida y")
plt.ylabel("|S[k]|")
plt.xlabel("Hz")
plt.stem(f, Sabs[0:int(N/2)], use_line_collection=True)

plt.show()
```

# Ahora lo planteo con la convolución lineal

```python
#planteo el delta de dirac
delta = np.zeros(N)
delta[0] = 1

#y[n]=(1/6) * ( x[n] - 2*x[n-1] + x[n-2] + 4*y[n-1] - 5*y[n-2] )
y[0] = (1/6)*delta[0]
y[1] = (1/6)*( delta[1] - 2*delta[1-1] + 4*y[1-1] )

for i in range(2,N):
    y[i] = (1/6) * (delta[i] - 2*delta[i-1] + delta[i-2] + 4*y[i-1] - 5*y[i-2])

#y[N-2] = (1/6) * ( delta[N-2] - 2*delta[N-3] + delta[N-4] + 4*y[N-2] - 5*y[N-4] )
#y[N-1] = (1/6) * ( delta[N-1] - 2*delta[N-2] + delta[N-3] + 4*y[N-1] - 5*y[N-3] )


#-------------------Grafico las señales
plt.subplots(2,1, figsize=(14,7))
#--------------------------
plt.subplot(2,1,1)
#plt.plot(t,U)
plt.stem(t, delta, use_line_collection=True)
plt.title("Señal de entrada Delta de dirac")
plt.ylabel("Amplitud")
#--------------------------
plt.subplot(2,1,2)
plt.plot(t,y)
plt.title("Respuesta al Impulso h(t)")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")

plt.show()
```

```python
#Respuesta al impulso
h = y
print(len(h))

#Convolución Lineal
y_conv = convolve(U,h, mode='full')

#------------------Grafico los espectros
plt.figure(figsize=(14,4))
plt.plot(y_conv)
plt.title("Señal de salida y_conv")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")

plt.show()
```

# Convolución Circular


Debido a como se calcula el producto interno en la convolución, no tengo la misma cantidad de muestras. Ahora debemos solucionar este problema, para ello empleamos la **convolución circular**.


Tener en cuenta que por las propiedades de la multiplicación de los espectros, debemos armar unos nuevas señales retocadas para hacer que coincidan la cantidad de muestras.

```python
#funciones acomodades para satisfacer las cantidades de muestras
U_prima = np.zeros(2*len(U)-1)
U_prima[:len(U)] = U

h_prima = np.zeros(2*len(h)-1)
h_prima[:len(h)] = h

print("tamaños U y h: ", len(U), len(h))
print("tamaños U y h primas: ", len(U_prima), len(h_prima))
```

```python
#-------------------Grafico las señales
plt.subplots(2,1, figsize=(14,7))
#--------------------------
plt.subplot(2,1,1)
#plt.plot(t,U)
plt.stem(U_prima, use_line_collection=True)
plt.title("Señal de entrada U_prima")
plt.ylabel("Amplitud")
#--------------------------
plt.subplot(2,1,2)
plt.plot(h_prima)
plt.title("Señal de salida h_prima")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")

plt.show()
```

```python
#-------------------Calculo el espectro
S_U_prima = fft(U_prima)
Sabs_U_prima = np.abs(S_U_prima)
f_U_prima = np.arange(0,fm/2,fm/len(U_prima))

S_h_prima = fft(h_prima)
Sabs_h_prima = np.abs(S_h_prima)
f_h_prima = np.arange(0,fm/2,fm/len(h_prima))

#------------------Grafico los espectros
plt.subplots(2,1, figsize=(14,7))
#---------------------------------
plt.subplot(2,1,1)
plt.title("Espectro de entrada U_prima")
plt.ylabel("|S[k]|")
plt.stem(f_U_prima[:-1], Sabs_U_prima[0:int(len(U_prima)/2)], use_line_collection=True)
#---------------------------------
plt.subplot(2,1,2)
plt.title("Espectro de salida h_prima")
plt.ylabel("|S[k]|")
plt.xlabel("Hz")
plt.stem(f_h_prima[:-1], Sabs_h_prima[0:int(len(h_prima)/2)], use_line_collection=True)

plt.show()
```

```python
#Multiplico las transformdas
multiplicacion = S_U_prima * S_h_prima #dominio frecuencial

y_inversa = ifft(multiplicacion).real #dominio temporal

#-------------------Grafico las señales
plt.subplots(2,1, figsize=(14,7))
#--------------------------
plt.subplot(2,1,1)
plt.plot(y_conv)
plt.title("Señal de salida y_conv")
plt.ylabel("Amplitud")
#--------------------------
plt.subplot(2,1,2)
plt.plot(y_inversa)
plt.title("Señal de salida y_inversa")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")

plt.show()
```

Gráficamente se ve que son iguales. Compruebo los tamaños:

```python
print("tamaños y_conv y y_inversa: ", len(y_conv), len(y_inversa))
```

**Ejercicio 8 modificado**

pb100.txt

- Respuesta al impulso de un filtro Pasa Bajos de 100 Hz
- fm = 1000 Hz

1. Análisis del filtro: temporal y frecuencial.
2. Usar el filtro
  - Señal de entrada x(t) --> plantear la suma de 2 senoidales de 20 Hz y 490 Hz.







```python
#-------------------Cargo los datos
pb100 = np.loadtxt("seNales/pb100.txt")
print("pb: ", pb100.shape)
```

```python
#Cantidad de muestras
N = len(pb100)

#Frecuencia de muestreo
fm = 1000
delta_t = 1/fm

#Eje temporal
t = np.arange(0,(N/fm),delta_t)

#-------------------Calculo el espectro
S = fft(pb100)
Sabs = np.abs(S)
f = np.arange(0,fm/2,fm/N)

#-------------------Grafico la señal
plt.subplots(2,1, figsize=(14,7))
#---------------------------------
plt.subplot(2,1,1)
plt.plot(t, pb100)
plt.title("Filtro pasa bajos")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")
#------------------Grafico el espectro
#--------------------------
plt.subplot(2,1,2)
plt.title("Espectro de filtro pb100")
plt.ylabel("|S[k]|")
plt.xlabel("Frecuencia")
plt.stem(f, Sabs[0:int(N/2)], use_line_collection=True)

plt.show()
```

```python
#Creamos la señal de entrada
Tm_x = 0.001 #1000 Hz
fm_x = 1/Tm_x
t_x = np.arange(0,1,1/fm_x)
N_x = len(t_x)

#Frecuencia de las señales que la componen
fs1 = 20 #está adentro del espectro del filtro
fs2 = 490 #está por fuera

#señal de entrada suma de senoidales
x = np.sin(2*np.pi*fs1*t_x) + (np.sin(2*np.pi*fs2*t_x))

#-------------------Calculo el espectro
S_x = fft(x)
Sabs_x = np.abs(S_x)
f_x = np.arange(0,fm_x/2,fm_x/N_x)

#-------------------Grafico la señal
plt.subplots(2,1, figsize=(14,7))
#---------------------------------
plt.subplot(2,1,1)
plt.plot(t_x, x)
plt.title("Señal de entrada x ")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")
#------------------Grafico el espectro
#--------------------------
plt.subplot(2,1,2)
plt.title("Espectro de entrada x")
plt.ylabel("|S[k]|")
plt.xlabel("Frecuencia")
plt.stem(f_x, Sabs_x[0:int(N_x/2)], use_line_collection=True)

plt.show()
```

```python
#Convolución Lineal
y_conv = convolve(x,pb100, mode='full')

N_y_conv = len(y_conv)
t_y_conv = np.arange(0, N_y_conv/fm, 1/fm)

#-------------------Calculo el espectro
S_y_conv = fft(y_conv)
Sabs_y_conv = np.abs(S_y_conv)
f_y_conv = np.arange(0,fm/2,fm/N_y_conv)


#-------------------Grafico la señal
plt.subplots(2,1, figsize=(14,7))
#---------------------------------
plt.subplot(2,1,1)
plt.plot(t_y_conv, y_conv)
plt.title("Señal de salida y_conv ")
plt.ylabel("Amplitud")
plt.xlabel("Tiempo")
#------------------Grafico el espectro
#--------------------------
plt.subplot(2,1,2)
plt.title("Espectro de salida y_conv")
plt.ylabel("|S[k]|")
plt.xlabel("Frecuencia")
plt.stem(f_y_conv[:-1], Sabs_y_conv[0:int(N_y_conv/2)], use_line_collection=True)

plt.show()
```

En **tiempo** se ve que limpia la señal del ruido que mete la senoidal de alta frecuencia y la deja como una senoidal linda.

En **frecuencia** se observa que se conserva el pico de 20 Hz (con un ruido inherente a aplicar el filtro) y que se elimina el pico de alta frecuencia porque el filtro solo deja pasar bajas frecuencias de menos de 100 Hz.
