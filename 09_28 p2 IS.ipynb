## Ejercicio 1

Señal aleatoria sintética

```python
# TP7 - Ej1

fm = 10000
t = np.arange(0, 1, 1/fm)
N = len(t)
#señal aleatoria - ruido uniforme
x = np.random.rand(len(t))-0.5
plt.plot(t,x)
plt.xlabel('tiempo (s.)')
plt.show()
```

```python
# Salida sistema

y = np.zeros(N)
y[0] = x[0]
y[1] = 0.3*y[0]+x[1]
y[2] = 0.3*y[1]-0.4*y[0]+x[2]
for n in range(3,N-1):
  y[n] = 0.3*y[n-1]-0.4*y[n-2]+0.2*y[n-3]+x[n] # De orden 3
plt.plot(t,y)
plt.show()
```

```python
ys = [x, y]
transf = [np.abs(fft(i)) for i in ys]
pos = 1
plt.figure(figsize=(15,8))

f = np.arange(0, fm/2, fm/N)

plt.subplot(211)

#plt.stem(f, transf[0][:int(N/2)], use_line_collection=True, linefmt="red", label="Espectro de x")
plt.title("Espectro de frecuencias de x")
plt.plot(f, transf[0][:int(N/2)])

plt.subplot(212)
#plt.stem(f, transf[1][:int(N/2)], use_line_collection=True, linefmt="red", label="Espectro de x")
plt.title("Espectro de frecuencias de y")
plt.plot(f, transf[1][:int(N/2)])
plt.show()
```

Es útil usar el ruido porque sigue la respuesta frecuencial.

Ver una salida que es ruido que provee de una entrada que es ruido tmb nos brinda poca información

```python
def lpc_autocorrelation(signal,p):

    n=signal.size
    R=np.correlate(signal,signal,'full') # 'full' quiere decir que la haga de - a + inf
    k=R[n]/R[n-1]
    a=np.array([-k])
    k=(R[n+1]+a*R[n])/(R[n-1]+a*R[n])

    for i in range(1,p):

        a1=np.zeros(i+1)
        a1[i]=-k
        for j in range(0,i):
            a1[j]=a[j]-k*a[i-1-j]
        del a
        a=a1
        k=(R[n+i+1]+np.sum(np.multiply(a,R[(n+i):(n-1):-1])))/(R[n-1]+np.sum(np.multiply(a,R[n:(n+i+1)])))
        del a1

    return(np.r_[1,a])
```

```python
r=np.correlate(y,y,'full')
plt.plot(r)
plt.title('Autocorrelación completa')
plt.xlabel('desplazamiento')
plt.ylabel('Autocorrelación')
plt.show()
```

Un clasificador muy simple de un clasificador de una señal perdiodica es ver si hay un pico periodico, porque sería que periodicamente se solapan y la autocorrelacion se solapan.

```python
r_parte = r[N-1:N+N]
plt.plot(r_parte)
plt.title('Mitad positiva')
plt.xlabel('desplazamiento')
plt.ylabel('Autocorrelación')
plt.show()
```

```python
plt.plot(r_parte[:100])
plt.title('Inicio de la mitad positiva')
plt.xlabel('desplazamiento')
plt.ylabel('Autocorrelación')
plt.show()
```

```python
pMax = 10 # Orden máximo que vamos a evaluar.
Vp = np.zeros(pMax)
Ip = np.zeros(pMax)
for p in range(1,pMax+1):
  a = lpc_autocorrelation(y,p)

  # error de prediccion final
  Ep=r_parte[0]+np.dot(r_parte[1:p+1],a[1:p+1])
  Vp[p-1]=Ep/r_parte[0];

  #criterio de Akaike
  Ip[p-1]=np.log(Ep)+2*p/(0.4*N)
```

```python
eje_p = range(1,pMax+1)
plt.plot(eje_p,Vp)
plt.xlabel('orden p')
plt.title('Vp')
plt.xticks(eje_p)
plt.show()
```

```python
plt.plot(eje_p,Ip)
plt.title('Ip')
plt.xlabel('orden p')
plt.xticks(eje_p)
plt.show()
```

En casos reales la planchada no va a estar tan clara pero va a haber una curva bastante clara, y en akaike no va a haber un mínimo global tan marcado. 

No siempre tengo que elegir el máximo, porque ahi me estoy haciendo un modelo de la señal con la que lo hice no del sistema.
